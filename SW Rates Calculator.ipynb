```python
import random
```

```python
cryst = 1000000 #number of crystals to use
totals = [0] * 28
collective = [0] * 13

shop = [0] * 5
```

```python
boxCryst = cryst #copy of crystals for boxes
while boxCryst >= 750: #can you afford a box
    boxCryst = boxCryst - 750 #buy a box
    for i in range(0,9): #10 things per box
        value = random.randint(1,10000)
        #trans x30 : .0001 
        if value == 1:
            totals[0] = totals[0] + 1
        #trans x20 : .0015 
        elif value <= 16:
            totals[1] = totals[1] + 1
        #trans x10 : .003
        elif value <= 46:
            totals[2] = totals[2] + 1
        #leg all-atrib : .0001
        elif value == 47:
            totals[3] = totals[3] + 1
        #leg water : .0001
        elif value <= 48:
            totals[4] = totals[4] + 1
        #leg fire : .0001
        elif value <= 49:
            totals[5] = totals[5] + 1
        #leg wind : .0001
        elif value <= 50:
            totals[6] = totals[6] + 1
        #leg scroll : .0005
        elif value <= 55:
            totals[7] = totals[7] + 1
        #leg x50 : .001
        elif value <= 65:
            totals[8] = totals[8] + 1
        #leg x40 : .002
        elif value <= 85:
            totals[9] = totals[9] + 1
        #leg x30 : .004
        elif value <= 125:
            totals[10] = totals[10] + 1
        #leg x20 : .005
        elif value <= 175:
            totals[11] = totals[11] + 1
        #leg x10 : .025
        elif value <= 425:
            totals[12] = totals[12] + 1
        #ld scroll : .0005
        elif value <= 430:
            totals[13] = totals[13] + 1
        #ld x25 : .001
        elif value <= 440:
            totals[14] = totals[14] + 1
        #ld x20 : .002
        elif value <= 460:
            totals[15] = totals[15] + 1
        #ld x15 : .004
        elif value <= 500:
            totals[16] = totals[16] + 1
        #ld x10 : .005
        elif value <= 550:
            totals[17] = totals[17] + 1
        #ld x5 : .025
        elif value <= 800:
            totals[18] = totals[18] + 1
        #water x3 : .01
        elif value <= 900:
            totals[19] = totals[19] + 1
        #water x2 : .02
        elif value <= 1100:
            totals[20] = totals[20] + 1
        #fire x3 : .01
        elif value <= 1200:
            totals[21] = totals[21] + 1
        #fire x2 : .02
        elif value <= 1400:
            totals[22] = totals[22] + 1
        #wind x3 : .01
        elif value <= 1500:
            totals[23] = totals[23] + 1
        #wind x2 : .02
        elif value <= 1700:
            totals[24] = totals[24] + 1
        #mystic x3 : .03
        elif value <= 2000:
            totals[25] = totals[25] + 1
        #mystic x2 : .15
        elif value <= 3500:
            totals[26] = totals[26] + 1
        #mystic x1 : .65
        else:
            totals[27] = totals[27] + 1
```

```python
#trans
collective[0] = totals[0] * 30 + totals[1] * 20 + totals[2] * 10

#leg all
collective[1] = totals[3]

#leg water
collective[2] = totals[4]

#leg fire
collective[3] = totals[5]

#leg wind
collective[4] = totals[6]

#leg
collective[5] = totals[7]

#leg pieces
collective[6] = totals[8] * 50 + totals[9] * 40 + totals[10] * 30 + totals[11] * 20 + totals[12] * 10

#ld
collective[7] = totals[13]

#ld pieces
collective[8] = totals[14] * 25 + totals[15] * 20 + totals[16] * 15 + totals[17] * 10 + totals[18] * 5

#water
collective[9] = totals[19] * 3 + totals[20] * 2

#fire
collective[10] = totals[21] * 3 + totals[22] * 2

#wind
collective[11] = totals[23] * 3 + totals[24] * 2

#mystic
collective[12] = totals[25] * 3 + totals[26] * 2 + totals[27]
```

```python
#shop refresh numbers from SWARFARM

#runes : .00015 (6 star leg rune of slot 1,3,5 or slot 2,4,6 with a not flat stat rune)
#mystic .0043
#leg pieces .0016
#ld pieces .0016
#stones .002
```

```python
shopCryst = cryst #copy of crystals for shop
while shopCryst > 3: #can you afford a refresh
    shopCryst = shopCryst-3 #refresh shop
    for i in range(0,9):
        value = random.randint(1,100000)
        
        #rune
        if value <=15:
            shop[0] = shop[0] + 1
        #mystic
        elif value <= 445:
            shop[1] = shop[1] + 1
        #leg pieces
        elif value <= 605:
            shop[2] = shop[2] + 1
        #ld pieces
        elif value <= 765:
            shop[3] = shop[3] + 1
        #stones
        elif value <= 965:
            shop[4] = shop[4] + 1
```
